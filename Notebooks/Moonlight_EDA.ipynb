```python
import pandas as pd
import numpy as np
```

```python
# reading the data
benin=pd.read_csv('benin-malanville.csv')
sierraleone=pd.read_csv('sierraleone-bumbuna.csv')
togo=pd.read_csv('togo-dapaong_qc.csv')
```

```python
negative_value_benin=benin[["GHI", "DNI", "DHI"]]<0
```

```python
benin[negative_value_benin]
```

```python
# removing the negative values is inconvinient so lets, make them positive
benin["GHI"]=benin["GHI"].abs()
benin["DNI"]=benin["DNI"].abs()
benin["DHI"]=benin["DHI"].abs()
```

```python
# same goes to the rest of data
# removing the negative values is inconvinient so lets, make them positive
togo["GHI"]=togo["GHI"].abs()
togo["DNI"]=togo["DNI"].abs()
togo["DHI"]=togo["DHI"].abs()
sierraleone["GHI"]=sierraleone["GHI"].abs()
sierraleone["DNI"]=sierraleone["DNI"].abs()
sierraleone["DHI"]=sierraleone["DHI"].abs()
```

```python
benin.isnull().sum()
```

```python
benin.dropna(axis="rows",inplace=True)
togo.dropna(axis="rows",inplace=True)
sierraleone.dropna(axis="rows",inplace=True)
```

```python
# Define the variables for which we want to create histograms
variables = ['GHI', 'DNI', 'DHI', 'WS', 'TModA', 'TModB', 'Tamb']

# Create a histogram for each variable
plt.figure(figsize=(15, 10))
for i, variable in enumerate(variables):
    plt.subplot(3, 3, i+1)
    benin[variable].hist(bins=20)
    plt.title(f'Histogram of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
